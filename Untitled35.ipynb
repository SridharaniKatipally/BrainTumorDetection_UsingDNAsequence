```python
# Step 1: Import Required Libraries
from Bio import SeqIO
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, Conv1D, MaxPooling1D, Flatten, Dense, Dropout
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns

# Step 2: Load FASTA Data
def load_fasta_files(file_paths):
    data = []
    for file_path in file_paths:
        for record in SeqIO.parse(file_path, "fasta"):
            data.append({
                "id": record.id,
                "sequence": str(record.seq),
                "description": record.description
            })
    return pd.DataFrame(data)

# List of FASTA file paths
file_paths = [
    "/content/sequence (5).fasta",
    "/content/sequence (6).fasta",
    "/content/sequence (7).fasta",
    "/content/sequence (8).fasta"
]

# Load the datasets
fasta_data = load_fasta_files(file_paths)
print(fasta_data.shape)
fasta_data.head()

# Step 3: Preprocess Data and Extract Features

# Generate random labels for each sequence
# In practice, replace this with actual labels
np.random.seed(42)
fasta_data['label'] = np.random.choice(['stage1', 'stage2', 'stage3'], size=len(fasta_data))

# Encode the sequences into numerical format
tokenizer = Tokenizer(num_words=5000, char_level=True)  # Consider adjusting the number of words
tokenizer.fit_on_texts(fasta_data['sequence'])
sequences = tokenizer.texts_to_sequences(fasta_data['sequence'])

# Pad the sequences to ensure uniform input size
max_sequence_length = 100  # Maximum length of text sequences
padded_sequences = pad_sequences(sequences, maxlen=max_sequence_length)

# Encode the labels (this example assumes a column 'label' for stages)
label_encoder = LabelEncoder()
encoded_labels = label_encoder.fit_transform(fasta_data['label'])  # Adjust 'label' column accordingly

# Split into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(padded_sequences, encoded_labels, test_size=0.2, random_state=42)

# Step 4: Build the Deep Learning Model

# Define the model
model = Sequential([
    Embedding(input_dim=len(tokenizer.word_index) + 1, output_dim=128, input_length=max_sequence_length),
    Conv1D(64, 5, activation='relu'),
    MaxPooling1D(pool_size=4),
    Conv1D(128, 5, activation='relu'),
    MaxPooling1D(pool_size=4),
    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(len(label_encoder.classes_), activation='softmax')
])

# Compile the model
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# Train the model
history = model.fit(X_train, y_train, epochs=10, validation_data=(X_test, y_test), batch_size=32)

# Print the summary of the model
model.summary()
```

```python
# Split into training and testing sets, including indices
X_train, X_test, y_train, y_test, train_index, test_index = train_test_split(
    padded_sequences, encoded_labels, fasta_data.index, test_size=0.2, random_state=42)

# Step 5: Make Predictions

predictions = model.predict(X_test)
predicted_labels = label_encoder.inverse_transform(predictions.argmax(axis=1))

# Create a DataFrame for the results using test_index to get original data
results = pd.DataFrame({
    'id': fasta_data.loc[test_index, 'id'],  # Accessing test indices from fasta_data
    'sequence': fasta_data.loc[test_index, 'sequence'],  # Accessing test sequences from fasta_data
    'predicted_label': predicted_labels
})
results.head()
```
